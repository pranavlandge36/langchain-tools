# Custom Tool

```python
from langchain_core.tools import tool
```

```python
@tool
def multiply(a:int,b:int)->int :
  'Multipy two numbers'
  return a*b
```

# Structured Tool

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
from typing import Required
class multiply_input(BaseModel):
  a:int=Field(Required,description='first number')
  b:int=Field(Required,description='second number')
```

```python
def multiply(a:int,b:int)->int:
  return a*b
```

```python
multipy_tool=StructuredTool.from_function(
    func=multiply,
    name='multiply',
    description='multiply two numbers',
    args_schema=multiply_input
)
```

```python
multipy_tool.invoke({'a':2,'b':3})
```

```python
print(multipy_tool.name)
print(multipy_tool.description)
print(multipy_tool.args)
print(multipy_tool.args_schema.model_json_schema())
```

# Base Model

```python
from langchain.tools import BaseTool
from typing import Type # Import Type
class Multipy_Tool(BaseTool):
  name: str = 'multiply'
  description: str = 'multiply two number'
  args_schema: Type[BaseModel]=multiply_input

  def _run(self, a:int,b:int)->int:
    return a*b
```

```python
tool=Multipy_Tool()
```

```python
tool.invoke({'a':2,'b':3})
```

# Toolkit

```python
from langchain_core.tools import tool
```

```python
@tool
def add(a:int,b:int)->int:
  'add two numbers'
  return a*b

@tool
def divide(a:int,b:int)->int:
  'divide two numbers'
  return a/b
```

```python
class MathToolkit:
  def get_tools(self):
    return [add,divide]
```

```python
toolkit=MathToolkit()
```

```python
tools=toolkit.get_tools()
```
